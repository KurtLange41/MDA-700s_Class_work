```python
class Perceptron(object):
    
    """Creating Perceptron classifier
    Parameters
    ----------------------
    eta:float
        learning rate(between 0.0 and 1.0)
    n_iter: int
        passovers the training dataset.
    random_state: int
        Random number generator weights for random weights
    
    Attributes:
    -----------------------
    w_:1d-array
        weights after fitting
    errors_: list 
        Number of missclassification (updates) on each epoch
    """
    
    def __init__(self,eta=0.01,n_iter=50,random_state=1):
        
        self.eta=eta
        self.n_iter=n_iter
        self.random_state=random_state
    
    def fit(self,X,y):
        rgen=np.random.RandomState(self.random_state)
        self.w_=rgen.normal(loc=0.0,scale=0.01,size=1+X.shape[1])  # add for the bias consideration
        #shape[1] because we need the attributes that is the columns
        self.errors_=[] # creating an empty list to fill after every epochs
        
        for i in range(self.n_iter):
            errors=0
            for xi,target in zip(X,y):
                update=self.eta*(target-self.predict(xi))
                # predict is the phi function
                self.w_[1:] +=update*xi
                #we start with 1 because 0th is the biased weight
                self.w_[0]+=update*1
                # because there is no x intercept with the biased weight
                errors+=int(update!=0.0)
                
            self.errors_.append(errors)
    
    def net_output(self,X):
        return (np.dot(X,self.w_[1:]) + self.w_[0])
    
    def predict(self, X):
        return np.where(self.net_output(X)>=0.0,1,-1)
        # if Z decision function >=0 give 1 else -1
    
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

Y = df_bank.iloc[0:4000, 16].values
Y = np.where(Y == 0, -1, 1)

X = df_bank.iloc[0:4000, [4,6,7,12]].values
```

```python
ppn = Perceptron(eta=0.0001, n_iter=10)

ppn.fit(X, Y)

plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')

# plt.savefig('images/02_07.png', dpi=300)
plt.show()
```
